```python
# ============================================================
# CAPSTONE FINAL — Multi-Ticker Idiosyncratic Strategy
# ============================================================
# Tickers: AAPL, MSFT, NVDA
# Hedge (default): SPY + Sector ETF per ticker
# Robustness: QQQ + Sector ETF per ticker
# Target: Next-day EXCESS return (stock - beta_mkt * market - beta_sect * sector)
# Model selection: choose model (LR or RF) that maximizes VALIDATION Sharpe (after costs)
# Portfolio: equal-weight across tickers each day (positions in {-1,0,+1})
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, average_precision_score

# ------------------------------------------------------------
# CONFIG
# ------------------------------------------------------------
TICKERS = ["AAPL", "MSFT", "NVDA"]   # you can append more later
START   = "1999-01-01"
END     = None                        # None => today

# Hedges
MARKET_DEFAULT = "SPY"                # default market proxy
MARKET_ALT     = "QQQ"                # robustness market proxy
SECTOR_MAP = {                        # default/fixed mapping (can be changed to dynamic)
    "AAPL":"XLK",
    "MSFT":"XLK",
    "NVDA":"XLK",   # NVDA is Tech; leave TSLA->XLY if you add back later
}

# Macros & proxies
MACRO = {
    "VIX": "^VIX",
    "HYG": "HYG",
    "LQD": "LQD",
    "UUP": "UUP",
    "IEF": "IEF",
    "TLT": "TLT",
    "TNX": "^TNX",
}

# Rolling beta settings
BETA_WINDOW = 60
BETA_MINPER = 20

# Splits
TEST_SIZE   = 0.20    # last 20% of dates for TEST
VAL_FRAC    = 0.20    # last 20% of TRAIN for validation

# Trading costs
BPS_PER_SIDE   = 0.0005     # 5 bps = 0.0005
SLIPPAGE_PER_SIDE = 0.0002  # 2 bps
PER_SIDE_COST  = BPS_PER_SIDE + SLIPPAGE_PER_SIDE

# Threshold / band grids for validation
THRESH_GRID = np.linspace(0.40, 0.60, 21)
BAND_GRID   = [0.00, 0.02, 0.05]

# Random Forest small grid (kept compact)
RF_GRID = [
    dict(n_estimators=400, max_depth=4,  min_samples_leaf=5, random_state=42),
    dict(n_estimators=600, max_depth=6,  min_samples_leaf=5, random_state=42),
    dict(n_estimators=800, max_depth=10, min_samples_leaf=3, random_state=42),
]

# ------------------------------------------------------------
# UTILITIES
# ------------------------------------------------------------
try:
    import yfinance as yf
except ImportError as e:
    raise ImportError("Please install yfinance:  !pip install yfinance") from e

def normalize_yf(df, tickers):
    """Return (Date x Ticker) prices table. Works for both groupings; prefers 'Adj Close'."""
    if isinstance(df.columns, pd.MultiIndex):
        lv0 = df.columns.get_level_values(0)
        lv1 = df.columns.get_level_values(1)
        if ('Adj Close' in lv0) or ('Close' in lv0):
            field_level, ticker_level = 0, 1
        elif ('Adj Close' in lv1) or ('Close' in lv1):
            field_level, ticker_level = 1, 0
        else:
            raise KeyError("Could not find price fields in MultiIndex columns.")
        if 'Adj Close' in df.columns.get_level_values(field_level):
            prices = df.xs('Adj Close', axis=1, level=field_level)
        elif 'Close' in df.columns.get_level_values(field_level):
            prices = df.xs('Close', axis=1, level=field_level)
        else:
            raise KeyError("Neither 'Adj Close' nor 'Close' found at detected field level.")
        present = [t for t in tickers if t in prices.columns]
        prices = prices[present].copy()
    else:
        # Single ticker
        if 'Adj Close' in df.columns:
            s = df['Adj Close'].rename(tickers[0])
        elif 'Close' in df.columns:
            s = df['Close'].rename(tickers[0])
        else:
            raise KeyError("Single-ticker data lacks 'Adj Close'/'Close'.")
        prices = s.to_frame()
    return prices.apply(pd.to_numeric, errors='coerce')

def rsi(series, window=14):
    delta = series.diff()
    up = delta.clip(lower=0.0)
    down = (-delta).clip(lower=0.0)
    roll_up = up.rolling(window, min_periods=window).mean()
    roll_down = down.rolling(window, min_periods=window).mean()
    rs = roll_up / (roll_down + 1e-12)
    out = 100 - 100 / (1 + rs)
    return out.replace([np.inf, -np.inf], np.nan)

def bb_percent_b(series, window=20, n_std=2):
    ma = series.rolling(window, min_periods=window).mean()
    sd = series.rolling(window, min_periods=window).std()
    upper = ma + n_std * sd
    lower = ma - n_std * sd
    return (series - lower) / (upper - lower)

def realized_vol_abs(ret, w):
    return ret.abs().rolling(w, min_periods=w).mean()

def rolling_beta(stock_ret, factor_ret, window=BETA_WINDOW, min_periods=BETA_MINPER):
    cov_w = stock_ret.rolling(window, min_periods=min_periods).cov(factor_ret)
    var_w = factor_ret.rolling(window, min_periods=min_periods).var()
    beta = cov_w / (var_w + 1e-12)
    # expanding fallback to avoid early NaN
    cov_e = stock_ret.expanding(min_periods=10).cov(factor_ret)
    var_e = factor_ret.expanding(min_periods=10).var()
    beta = beta.combine_first(cov_e / (var_e + 1e-12))
    # gentle clipping to avoid explosions
    return beta.clip(lower=-3, upper=3)

def perf_stats(returns, trading_days=252):
    r = pd.Series(returns).dropna()
    if len(r) == 0:
        return dict(CAGR=np.nan, Sharpe=np.nan, MaxDD=np.nan)
    cum = (1 + r).cumprod()
    years = len(r)/trading_days
    cagr = cum.iloc[-1]**(1/years) - 1 if years>0 else np.nan
    sharpe = np.sqrt(trading_days) * r.mean() / (r.std() + 1e-12)
    dd = cum/cum.cummax() - 1
    return dict(CAGR=cagr, Sharpe=sharpe, MaxDD=dd.min())

def txn_costs_from_positions(pos_series, per_side_cost=PER_SIDE_COST):
    """Cost in returns space given position changes (pos in {-1,0,+1})."""
    pos = pos_series.fillna(0).astype(float)
    turn = pos.diff().abs()  # change of 1 => one side; change of 2 => two sides
    return -per_side_cost * turn

def fit_and_choose_model(X_tr, y_tr, X_val, y_val, val_index, val_excess):
    """Train LR and RF candidates; choose by validation Sharpe (after costs)."""
    scaler = StandardScaler()
    X_tr_sc = scaler.fit_transform(X_tr)
    X_val_sc = scaler.transform(X_val)

    candidates = []

    # Logistic Regression (simple)
    for C in [0.1, 1.0, 3.0, 10.0]:
        lr = LogisticRegression(class_weight='balanced', C=C, max_iter=2000, n_jobs=None)
        lr.fit(X_tr_sc, y_tr)
        val_proba = lr.predict_proba(X_val_sc)[:,1]
        # tune threshold+band by Sharpe
        best_tuple, best_sharpe = None, -np.inf
        for th in THRESH_GRID:
            for band in BAND_GRID:
                pos = np.where(val_proba >= th + band, 1,
                        np.where(val_proba <= th - band, -1, 0))
                r = pd.Series(pos, index=val_index) * val_excess
                r = r.dropna()
                costs = txn_costs_from_positions(pd.Series(pos, index=val_index))
                rr = (r + costs).dropna()
                if rr.std() > 0 and len(rr) > 30:
                    sh = np.sqrt(252) * rr.mean() / rr.std()
                    if sh > best_sharpe:
                        best_sharpe = sh
                        best_tuple = ("LR", C, th, band, val_proba)
        if best_tuple:
            candidates.append(dict(model_name="LR", C=best_tuple[1], thr=best_tuple[2],
                                   band=best_tuple[3], proba=best_tuple[4], scaler=scaler,
                                   estimator=lr, val_sharpe=best_sharpe))

    # Random Forest (compact grid)
    for cfg in RF_GRID:
        rf = RandomForestClassifier(**cfg, class_weight="balanced")
        rf.fit(X_tr, y_tr)  # trees handle unscaled inputs fine
        val_proba = rf.predict_proba(X_val)[:,1]
        best_tuple, best_sharpe = None, -np.inf
        for th in THRESH_GRID:
            for band in BAND_GRID:
                pos = np.where(val_proba >= th + band, 1,
                        np.where(val_proba <= th - band, -1, 0))
                r = pd.Series(pos, index=val_index) * val_excess
                r = r.dropna()
                costs = txn_costs_from_positions(pd.Series(pos, index=val_index))
                rr = (r + costs).dropna()
                if rr.std() > 0 and len(rr) > 30:
                    sh = np.sqrt(252) * rr.mean() / rr.std()
                    if sh > best_sharpe:
                        best_sharpe = sh
                        best_tuple = ("RF", cfg, th, band, val_proba)
        if best_tuple:
            candidates.append(dict(model_name="RF", params=cfg, thr=best_tuple[2],
                                   band=best_tuple[3], proba=best_tuple[4], scaler=None,
                                   estimator=rf, val_sharpe=best_sharpe))

    if not candidates:
        raise RuntimeError("No viable model candidate found on validation.")

    # Choose best by validation Sharpe
    best = max(candidates, key=lambda d: d["val_sharpe"])
    return best

def model_probabilities(best, X_fit, X_test):
    """Apply scaling if needed; return (proba_fit, proba_test)."""
    if best["model_name"] == "LR":
        sc = best["scaler"]
        X_fit_sc  = sc.transform(X_fit)
        X_test_sc = sc.transform(X_test)
        return best["estimator"].predict_proba(X_fit_sc)[:,1], best["estimator"].predict_proba(X_test_sc)[:,1]
    else:
        return best["estimator"].predict_proba(X_fit)[:,1], best["estimator"].predict_proba(X_test)[:,1]

# ------------------------------------------------------------
# DATA DOWNLOAD
# ------------------------------------------------------------
ALL = sorted(set(TICKERS + [MARKET_DEFAULT, MARKET_ALT] + list(SECTOR_MAP.values()) + list(MACRO.values())))
raw = yf.download(ALL, start=START, end=END, auto_adjust=False, progress=False)
prices = normalize_yf(raw, ALL).sort_index().dropna(how='all')
rets = prices.pct_change()

# ------------------------------------------------------------
# FEATURE BUILD (functions use global rets/prices)
# ------------------------------------------------------------
def build_features_for_ticker(tkr, market_symbol, sector_symbol):
    """Return a DataFrame with features & target for one ticker and one market proxy."""
    px = prices[tkr]
    rS = rets[tkr]
    rM = rets[market_symbol]
    rSec = rets[sector_symbol] if sector_symbol in rets else None

    df = pd.DataFrame(index=prices.index)
    df['px']      = px
    df['ret_s']   = rS
    df['ret_m']   = rM
    if rSec is not None:
        df['ret_sec'] = rSec

    # Technicals / vol
    df['ret_s_lag1'] = rS.shift(1)
    df['ret_m_lag1'] = rM.shift(1)
    if rSec is not None:
        df['ret_sec_lag1'] = rSec.shift(1)

    df['rv5_s']  = realized_vol_abs(rS, 5)
    df['rv22_s'] = realized_vol_abs(rS, 22)
    df['rv5_m']  = realized_vol_abs(rM, 5)
    df['rv22_m'] = realized_vol_abs(rM, 22)
    df['ma10']   = px.rolling(10, min_periods=10).mean()
    df['ma30']   = px.rolling(30, min_periods=30).mean()
    df['ma_spread'] = df['ma10'] - df['ma30']
    df['rsi14']  = rsi(px, 14)
    df['bbp20']  = bb_percent_b(px, 20, 2)

    # Macro proxies
    if MACRO["VIX"] in prices:
        df['vix']  = prices[MACRO["VIX"]]
        df['dvix'] = df['vix'].pct_change()
    if "HYG" in rets and "LQD" in rets:
        df['credit_spread'] = rets["HYG"] - rets["LQD"]
    for m in ["UUP","IEF","TLT"]:
        if m in rets: df[f"ret_{m}_lag1"] = rets[m].shift(1)
    if MACRO["TNX"] in prices:
        df['d_tnx'] = prices[MACRO["TNX"]].pct_change()

    # Next-day hedged target
    df['ret_s_next'] = rS.shift(-1)
    df['ret_m_next'] = rM.shift(-1)

    beta_m = rolling_beta(df['ret_s'], df['ret_m'], window=BETA_WINDOW, min_periods=BETA_MINPER)
    if rSec is not None:
        df['ret_sec_next'] = rSec.shift(-1)
        beta_s = rolling_beta(df['ret_s'], df['ret_sec'], window=BETA_WINDOW, min_periods=BETA_MINPER)
        df['excess_ret_next'] = df['ret_s_next'] - (beta_m * df['ret_m_next'] + beta_s * df['ret_sec_next'])
    else:
        df['excess_ret_next'] = df['ret_s_next'] - (beta_m * df['ret_m_next'])

    df['y_excess_up'] = (df['excess_ret_next'] > 0).astype(int)

    drop_cols = ['px','ret_s_next','ret_m_next','ret_sec_next'] if 'ret_sec_next' in df.columns else ['px','ret_s_next','ret_m_next']
    df = df.dropna().copy()
    features = [c for c in df.columns if c not in (drop_cols + ['excess_ret_next','y_excess_up'])]
    target   = 'y_excess_up'
    return df, features, target

def run_single_ticker(tkr, market_symbol, sector_symbol):
    """Train/validate/test for one ticker; return dict with metrics and series."""
    df, features, target = build_features_for_ticker(tkr, market_symbol, sector_symbol)
    X = df[features].values
    y = df[target].values

    # time split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, shuffle=False)
    idx_train = df.index[:len(X_train)]
    idx_test  = df.index[len(X_train):]

    # split validation from tail of TRAIN
    cut = int(len(X_train)*(1-VAL_FRAC))
    X_tr, X_val = X_train[:cut], X_train[cut:]
    y_tr, y_val = y_train[:cut], y_train[cut:]
    idx_val = idx_train[cut:]

    # validation excess returns (aligned to idx_val)
    val_excess = df.loc[idx_val, 'excess_ret_next']

    # pick best model by validation Sharpe
    best = fit_and_choose_model(X_tr, y_tr, X_val, y_val, idx_val, val_excess)

    # probabilities on full fit segment (train) and test
    p_fit, p_test = model_probabilities(best, X_train, X_test)

    # AUC/PR on TEST (diagnostics)
    auc  = roc_auc_score(y_test, p_test)
    prc  = average_precision_score(y_test, p_test)

    # Build TEST positions using chosen threshold/band
    th, band = best["thr"], best["band"]
    pos_test = np.where(p_test >= th + band, 1, np.where(p_test <= th - band, -1, 0))
    pos_test = pd.Series(pos_test, index=idx_test)

    # Strategy returns: hedged excess
    strat_r = pos_test * df.loc[idx_test, 'excess_ret_next']
    costs   = txn_costs_from_positions(pos_test, PER_SIDE_COST)
    strat_net = (strat_r + costs).dropna()

    # Perf
    perf = perf_stats(strat_net)

    out = dict(
        ticker=tkr, market=market_symbol, sector=sector_symbol,
        best_model=best["model_name"],
        best_params=(best.get("C") if best["model_name"]=="LR" else best.get("params")),
        threshold=float(th), band=float(band),
        val_sharpe=float(best["val_sharpe"]),
        test_auc=float(auc), test_pr=float(prc),
        strat_returns=strat_net,  # pd.Series
        pos=pos_test,            # pd.Series
        df=df, features=features
    )
    return out

def combine_portfolio(results):
    """Equal-weight combination across tickers per day (only among non-NaN names)."""
    pnl = pd.DataFrame({f"{r['ticker']}": r['strat_returns'] for r in results})
    port = pnl.mean(axis=1, skipna=True)
    return port, {r['ticker']: perf_stats(r['strat_returns']) for r in results}, perf_stats(port)

# ------------------------------------------------------------
# RUN: DEFAULT HEDGE (SPY + Sector) AND ROBUSTNESS (QQQ + Sector)
# ------------------------------------------------------------
def run_suite(market_symbol):
    results = []
    for t in TICKERS:
        sector = SECTOR_MAP.get(t, None)
        if sector is None or sector not in prices.columns:
            # fallback: dynamic sector by absolute correlation over long lookback
            sector_candidates = [s for s in ["XLK","XLY","XLF","XLE","XLP","XLI","XLB","XLV","XLU","XLRE","XLC"] if s in prices.columns]
            if sector_candidates:
                cors = {s: rets[t].corr(rets[s]) for s in sector_candidates}
                sector = max(cors, key=lambda k: abs(cors[k]))
        results.append(run_single_ticker(t, market_symbol, sector))
    port, per_tkr, port_perf = combine_portfolio(results)
    return results, port, per_tkr, port_perf

print("Running default hedge: SPY + sector ...")
res_spy, port_spy, per_spy, port_spy_perf = run_suite(MARKET_DEFAULT)

print("Running robustness hedge: QQQ + sector ...")
res_qqq, port_qqq, per_qqq, port_qqq_perf = run_suite(MARKET_ALT)

# ------------------------------------------------------------
# REPORT
# ------------------------------------------------------------
def pretty_perf(p):
    return f"CAGR={p['CAGR']:.3%} | Sharpe={p['Sharpe']:.2f} | MaxDD={p['MaxDD']:.1%}"

print("\n=== PER-TICKER (SPY+Sector) ===")
for r in res_spy:
    stats = perf_stats(r['strat_returns'])
    print(f"{r['ticker']:>5s}  [{r['best_model']}, thr={r['threshold']:.2f}, band={r['band']:.2f}]  "
          f"AUC={r['test_auc']:.3f}, PR={r['test_pr']:.3f}  ||  {pretty_perf(stats)}")

print("\n=== PORTFOLIO (Equal-Weight, SPY+Sector) ===")
print(pretty_perf(port_spy_perf))

print("\n=== PER-TICKER (QQQ+Sector) ===")
for r in res_qqq:
    stats = perf_stats(r['strat_returns'])
    print(f"{r['ticker']:>5s}  [{r['best_model']}, thr={r['threshold']:.2f}, band={r['band']:.2f}]  "
          f"AUC={r['test_auc']:.3f}, PR={r['test_pr']:.3f}  ||  {pretty_perf(stats)}")

print("\n=== PORTFOLIO (Equal-Weight, QQQ+Sector) ===")
print(pretty_perf(port_qqq_perf))

# ------------------------------------------------------------
# PLOTS
# ------------------------------------------------------------
plt.figure(figsize=(12,5))
(1 + port_spy).cumprod().plot(label="Portfolio: SPY + Sector", lw=2)
(1 + port_qqq).cumprod().plot(label="Portfolio: QQQ + Sector", lw=2, alpha=0.8)
plt.title("Cumulative Returns — Portfolio (TEST period portion inside series)")
plt.legend()
plt.tight_layout()
plt.savefig("./figures/port_v1.png", dpi=300, bbox_inches="tight") 
plt.show()

# Optional: per-ticker curves (default hedge)
plt.figure(figsize=(12,6))
for r in res_spy:
    (1 + r['strat_returns']).cumprod().plot(label=f"{r['ticker']} (SPY+Sector)", alpha=0.8)
plt.title("Cumulative Returns — Per Ticker (SPY+Sector)")
plt.legend()
plt.tight_layout()
plt.savefig("./figures/stock_v1.png", dpi=300, bbox_inches="tight") 
plt.show()
```
