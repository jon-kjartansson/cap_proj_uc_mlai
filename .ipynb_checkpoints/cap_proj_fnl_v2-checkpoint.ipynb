```python
# ============================================================
# CAPSTONE FINAL — Flexible Multi-Ticker Idiosyncratic Strategy
# Works for any list of S&P 1500 tickers (and most US tickers on Yahoo).
# ============================================================

import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, average_precision_score

# -----------------------------
# USER CONFIG
# -----------------------------
# Put ANY list of tickers here (S&P 1500 or otherwise).
TICKERS = ["AAPL", "MSFT", "NVDA"]  # Can be any US listed stock

# Date range: pick an overlap that exists for your list + ETFs
START   = "2000-01-01" # If list of tickers includes a newly listed name, specify a later START date
END     = None  # None => today

# Hedge market proxies (we will run both for robustness)
MARKET_DEFAULT = "SPY" # E.g. SPY or ^GSPC
MARKET_ALT     = "QQQ"

# Sector ETFs to consider (SPDRs)
SECTOR_ETFS = ["XLK","XLY","XLF","XLE","XLP","XLI","XLB","XLV","XLU","XLRE","XLC"]

# Size ETFs to consider
SIZE_MID = "MDY"  # S&P 400 mid-cap
SIZE_SML = "IJR"  # S&P 600 small-cap

# Optional fixed sector overrides (leave empty to let code auto-pick by correlation)
SECTOR_OVERRIDE = {
    # "AAPL": "XLK",
    # "MSFT": "XLK",
}

# Optional fixed size overrides: "mid" or "small", or None
SIZE_OVERRIDE = {
    # "TRU": "mid",
    # "STRL": "small",
}

# Rolling beta settings
BETA_WINDOW = 60 # Consider changing to 90 for noisy small caps
BETA_MINPER = 20

# Splits
TEST_SIZE = 0.20       # last 20% as TEST
VAL_FRAC  = 0.20       # last 20% of TRAIN as validation

# Trading costs (per side): 5 bps + 2 bps slippage
BPS_PER_SIDE        = 5e-4 # Consider increasing cost if churn is high
SLIPPAGE_PER_SIDE   = 2e-4
PER_SIDE_COST       = BPS_PER_SIDE + SLIPPAGE_PER_SIDE

# Threshold/band search space (Sharpe on validation)
THRESH_GRID = np.linspace(0.40, 0.60, 21)
BAND_GRID   = [0.00, 0.02, 0.05] # Consider increasing values if churn is high

# Random Forest compact configs (to keep runtime reasonable)
RF_GRID = [
    dict(n_estimators=400, max_depth=4,  min_samples_leaf=5, random_state=42),
    dict(n_estimators=600, max_depth=6,  min_samples_leaf=5, random_state=42),
    dict(n_estimators=800, max_depth=10, min_samples_leaf=3, random_state=42),
]

# -----------------------------
# LIBS & DOWNLOAD
# -----------------------------
try:
    import yfinance as yf
except ImportError as e:
    raise ImportError("Please install yfinance (e.g., !pip install yfinance)") from e

ALL_DOWNLOAD = sorted(set(
    TICKERS + [MARKET_DEFAULT, MARKET_ALT, SIZE_MID, SIZE_SML] + SECTOR_ETFS +
    ["^VIX","HYG","LQD","UUP","IEF","TLT","^TNX"]  # optional macro proxies for features
))

raw = yf.download(ALL_DOWNLOAD, start=START, end=END, auto_adjust=False, progress=False)

def normalize_yf(df, tickers):
    """Return (Date x Ticker) prices table. Works for both MultiIndex layouts; prefers 'Adj Close'."""
    if isinstance(df.columns, pd.MultiIndex):
        lv0 = df.columns.get_level_values(0)
        lv1 = df.columns.get_level_values(1)
        if ('Adj Close' in lv0) or ('Close' in lv0):
            field_level, ticker_level = 0, 1
        elif ('Adj Close' in lv1) or ('Close' in lv1):
            field_level, ticker_level = 1, 0
        else:
            raise KeyError("Could not find price fields in MultiIndex columns.")
        if 'Adj Close' in df.columns.get_level_values(field_level):
            prices = df.xs('Adj Close', axis=1, level=field_level)
        elif 'Close' in df.columns.get_level_values(field_level):
            prices = df.xs('Close', axis=1, level=field_level)
        else:
            raise KeyError("Neither 'Adj Close' nor 'Close' found at detected field level.")
        present = [t for t in tickers if t in prices.columns]
        prices = prices[present].copy()
    else:
        # Single ticker layout
        if 'Adj Close' in df.columns:
            s = df['Adj Close'].rename(tickers[0])
        elif 'Close' in df.columns:
            s = df['Close'].rename(tickers[0])
        else:
            raise KeyError("Single-ticker data lacks 'Adj Close'/'Close'.")
        prices = s.to_frame()
    return prices.apply(pd.to_numeric, errors='coerce')

prices = normalize_yf(raw, ALL_DOWNLOAD).sort_index().dropna(how='all')
rets    = prices.pct_change()

# -----------------------------
# HELPERS
# -----------------------------
def rsi(series, window=14):
    delta = series.diff()
    up = delta.clip(lower=0.0)
    down = (-delta).clip(lower=0.0)
    roll_up = up.rolling(window, min_periods=window).mean()
    roll_down = down.rolling(window, min_periods=window).mean()
    rs = roll_up / (roll_down + 1e-12)
    out = 100 - 100 / (1 + rs)
    return out.replace([np.inf, -np.inf], np.nan)

def bb_percent_b(series, window=20, n_std=2):
    ma = series.rolling(window, min_periods=window).mean()
    sd = series.rolling(window, min_periods=window).std()
    upper = ma + n_std * sd
    lower = ma - n_std * sd
    return (series - lower) / (upper - lower)

def realized_vol_abs(ret, w):
    return ret.abs().rolling(w, min_periods=w).mean()

def rolling_beta(y, x, window=BETA_WINDOW, min_periods=BETA_MINPER):
    cov_w = y.rolling(window, min_periods=min_periods).cov(x)
    var_w = x.rolling(window, min_periods=min_periods).var()
    beta = cov_w / (var_w + 1e-12)
    # expanding fallback to fill early NaNs
    cov_e = y.expanding(min_periods=10).cov(x)
    var_e = x.expanding(min_periods=10).var()
    beta = beta.combine_first(cov_e / (var_e + 1e-12))
    return beta.clip(-3, 3)

def perf_stats(returns, trading_days=252):
    r = pd.Series(returns).dropna()
    if len(r) == 0:
        return dict(CAGR=np.nan, Sharpe=np.nan, MaxDD=np.nan)
    cum = (1 + r).cumprod()
    years = len(r)/trading_days
    cagr = cum.iloc[-1]**(1/years) - 1 if years > 0 else np.nan
    sharpe = np.sqrt(trading_days) * r.mean() / (r.std() + 1e-12)
    dd = cum / cum.cummax() - 1
    return dict(CAGR=cagr, Sharpe=sharpe, MaxDD=dd.min())

def txn_costs_from_positions(pos_series, per_side_cost=PER_SIDE_COST):
    pos = pos_series.fillna(0).astype(float)
    turn = pos.diff().abs()  # change of 1 => one side; change of 2 => two sides
    return -per_side_cost * turn

# --- Size factor selection (mid/small) with multiple fallbacks ---
_market_cap_cache = {}

def guess_size_bucket(ticker):
    # 1) explicit override
    if ticker in SIZE_OVERRIDE:
        return SIZE_OVERRIDE[ticker]

    # 2) try yfinance fast_info/info to get market cap (may fail silently)
    try:
        if ticker not in _market_cap_cache:
            _market_cap_cache[ticker] = yf.Ticker(ticker).fast_info.get("market_cap", None)
        mc = _market_cap_cache[ticker]
        if mc is None:
            info = yf.Ticker(ticker).get_info()
            mc = info.get("marketCap", None)
            _market_cap_cache[ticker] = mc
        if mc:
            # Rough buckets (can tweak):
            # small < $2.5B, mid < $13B, else large
            if mc < 2.5e9:  return "small"
            if mc < 13e9:   return "mid"
            return None  # large => no explicit size factor
    except Exception:
        pass

    # 3) correlation fallback: compare corr with SPY vs MDY/IJR (full sample)
    if (ticker in rets.columns and
        SIZE_MID in rets.columns and SIZE_SML in rets.columns and
        MARKET_DEFAULT in rets.columns):
        c_spy = rets[ticker].corr(rets[MARKET_DEFAULT])
        c_mdy = rets[ticker].corr(rets[SIZE_MID])
        c_ijr = rets[ticker].corr(rets[SIZE_SML])
        best = max([("mid", c_mdy), ("small", c_ijr), ("large", c_spy)], key=lambda kv: abs(kv[1]))
        # If SPY correlation is best by a margin, treat as large
        if best[0] == "large":
            return None
        else:
            return best[0]
    return None

def pick_sector_etf_for_ticker(ticker):
    # 1) explicit override
    if ticker in SECTOR_OVERRIDE and SECTOR_OVERRIDE[ticker] in prices.columns:
        return SECTOR_OVERRIDE[ticker]

    # 2) correlation across available sector ETFs
    candidates = [s for s in SECTOR_ETFS if s in rets.columns]
    if not candidates:
        return None
    cors = {s: rets[ticker].corr(rets[s]) for s in candidates if ticker in rets.columns}
    # pick highest absolute correlation
    if not cors:
        return None
    return max(cors, key=lambda k: abs(cors[k]))

# -----------------------------
# FEATURE & TARGET BUILDER
# -----------------------------
def build_features_for_ticker(tkr, market_symbol):
    """Return df(features+target), feature list."""
    if tkr not in prices.columns or market_symbol not in prices.columns:
        raise KeyError(f"Missing price data for {tkr} or market {market_symbol}")

    px  = prices[tkr]
    rS  = rets[tkr]
    rM  = rets[market_symbol]

    # sector & size selections
    sector = pick_sector_etf_for_ticker(tkr)
    rSec = rets[sector] if (sector and sector in rets.columns) else None

    size_bucket = guess_size_bucket(tkr)  # "mid" / "small" / None
    if size_bucket == "mid" and SIZE_MID in rets:
        size_symbol = SIZE_MID
    elif size_bucket == "small" and SIZE_SML in rets:
        size_symbol = SIZE_SML
    else:
        size_symbol = None
    rZ = rets[size_symbol] if size_symbol else None

    # build frame
    df = pd.DataFrame(index=prices.index)
    df['px']    = px
    df['ret_s'] = rS
    df['ret_m'] = rM
    if rSec is not None:
        df['ret_sec'] = rSec
    if rZ is not None:
        df['ret_size'] = rZ

    # technicals / vol
    df['ret_s_lag1'] = rS.shift(1)
    df['ret_m_lag1'] = rM.shift(1)
    if rSec is not None:
        df['ret_sec_lag1'] = rSec.shift(1)
    if rZ is not None:
        df['ret_size_lag1'] = rZ.shift(1)

    df['rv5_s']   = realized_vol_abs(rS, 5)
    df['rv22_s']  = realized_vol_abs(rS, 22)
    df['rv5_m']   = realized_vol_abs(rM, 5)
    df['rv22_m']  = realized_vol_abs(rM, 22)
    df['ma10']    = px.rolling(10, min_periods=10).mean()
    df['ma30']    = px.rolling(30, min_periods=30).mean()
    df['ma_spread']= df['ma10'] - df['ma30']
    df['rsi14']   = rsi(px, 14)
    df['bbp20']   = bb_percent_b(px, 20, 2)

    # macro proxies (optional features; safe if missing)
    if "^VIX" in prices.columns:
        df['vix']  = prices["^VIX"]
        df['dvix'] = df['vix'].pct_change()
    if "HYG" in rets.columns and "LQD" in rets.columns:
        df['credit_spread'] = rets["HYG"] - rets["LQD"]
    for m in ["UUP","IEF","TLT"]:
        if m in rets.columns:
            df[f"ret_{m}_lag1"] = rets[m].shift(1)
    if "^TNX" in prices.columns:
        df['d_tnx'] = prices["^TNX"].pct_change()

    # targets and hedging (sequential residualization)
    df['ret_s_next'] = df['ret_s'].shift(-1)
    df['ret_m_next'] = df['ret_m'].shift(-1)
    if rSec is not None:
        df['ret_sec_next'] = df['ret_sec'].shift(-1)
    if rZ is not None:
        df['ret_size_next'] = df['ret_size'].shift(-1)

    # betas: market -> sector (on residual) -> size (on residual)
    beta_m = rolling_beta(df['ret_s'], df['ret_m'])
    resid1 = df['ret_s'] - beta_m * df['ret_m']

    beta_s = None
    if rSec is not None:
        beta_s = rolling_beta(resid1, df['ret_sec'])
        resid2 = resid1 - beta_s * df['ret_sec']
    else:
        resid2 = resid1

    beta_z = None
    if rZ is not None:
        beta_z = rolling_beta(resid2, df['ret_size'])

    hedge_next = beta_m * df['ret_m_next']
    if beta_s is not None: hedge_next = hedge_next + beta_s * df.get('ret_sec_next')
    if beta_z is not None: hedge_next = hedge_next + beta_z * df.get('ret_size_next')

    df['excess_ret_next'] = df['ret_s_next'] - hedge_next
    df['y_excess_up'] = (df['excess_ret_next'] > 0).astype(int)

    # finalize
    drop_cols = ['px','ret_s_next','ret_m_next']
    if 'ret_sec_next' in df:  drop_cols.append('ret_sec_next')
    if 'ret_size_next' in df: drop_cols.append('ret_size_next')

    df = df.dropna().copy()
    features = [c for c in df.columns if c not in (drop_cols + ['excess_ret_next','y_excess_up'])]
    meta = dict(sector=sector, size=size_symbol, market=market_symbol)
    return df, features, meta

# -----------------------------
# MODEL FIT & SELECTION
# -----------------------------
def fit_and_choose_model(X_tr, y_tr, X_val, y_val, val_index, val_excess):
    scaler = StandardScaler()
    X_tr_sc = scaler.fit_transform(X_tr)
    X_val_sc = scaler.transform(X_val)

    candidates = []

    # Logistic Regression
    for C in [0.1, 1.0, 3.0, 10.0]:
        lr = LogisticRegression(class_weight='balanced', C=C, max_iter=2000)
        lr.fit(X_tr_sc, y_tr)
        val_proba = lr.predict_proba(X_val_sc)[:,1]
        best_tuple, best_sharpe = None, -np.inf
        for th in THRESH_GRID:
            for band in BAND_GRID:
                pos = np.where(val_proba >= th + band, 1,
                               np.where(val_proba <= th - band, -1, 0))
                r = pd.Series(pos, index=val_index) * val_excess
                costs = txn_costs_from_positions(pd.Series(pos, index=val_index))
                rr = (r + costs).dropna()
                if rr.std() > 0 and len(rr) > 30:
                    sh = np.sqrt(252) * rr.mean() / rr.std()
                    if sh > best_sharpe:
                        best_sharpe = sh
                        best_tuple = ("LR", C, th, band, val_proba)
        if best_tuple:
            candidates.append(dict(model="LR", C=best_tuple[1], thr=best_tuple[2],
                                   band=best_tuple[3], proba=best_tuple[4],
                                   scaler=scaler, est=lr, val_sharpe=best_sharpe))

    # Random Forest (unscaled)
    for cfg in RF_GRID:
        rf = RandomForestClassifier(**cfg, class_weight="balanced")
        rf.fit(X_tr, y_tr)
        val_proba = rf.predict_proba(X_val)[:,1]
        best_tuple, best_sharpe = None, -np.inf
        for th in THRESH_GRID:
            for band in BAND_GRID:
                pos = np.where(val_proba >= th + band, 1,
                               np.where(val_proba <= th - band, -1, 0))
                r = pd.Series(pos, index=val_index) * val_excess
                costs = txn_costs_from_positions(pd.Series(pos, index=val_index))
                rr = (r + costs).dropna()
                if rr.std() > 0 and len(rr) > 30:
                    sh = np.sqrt(252) * rr.mean() / rr.std()
                    if sh > best_sharpe:
                        best_sharpe = sh
                        best_tuple = ("RF", cfg, th, band, val_proba)
        if best_tuple:
            candidates.append(dict(model="RF", params=cfg, thr=best_tuple[2],
                                   band=best_tuple[3], proba=best_tuple[4],
                                   scaler=None, est=rf, val_sharpe=best_sharpe))

    if not candidates:
        raise RuntimeError("No viable model candidate on validation.")

    return max(candidates, key=lambda d: d["val_sharpe"])

def model_probabilities(best, X_fit, X_test):
    if best["model"] == "LR":
        sc = best["scaler"]
        return (best["est"].predict_proba(sc.transform(X_fit))[:,1],
                best["est"].predict_proba(sc.transform(X_test))[:,1])
    else:
        return best["est"].predict_proba(X_fit)[:,1], best["est"].predict_proba(X_test)[:,1]

# -----------------------------
# SINGLE-TICKER PIPELINE
# -----------------------------
def run_single_ticker(tkr, market_symbol):
    df, features, meta = build_features_for_ticker(tkr, market_symbol)
    X = df[features].values
    y = df['y_excess_up'].values

    # time-ordered split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, shuffle=False)
    idx_train = df.index[:len(X_train)]
    idx_test  = df.index[len(X_train):]

    # validation slice (tail of TRAIN)
    cut = int(len(X_train) * (1 - VAL_FRAC))
    X_tr, X_val = X_train[:cut], X_train[cut:]
    y_tr, y_val = y_train[:cut], y_train[cut:]
    idx_val = idx_train[cut:]

    val_excess = df.loc[idx_val, 'excess_ret_next']

    # choose model by validation Sharpe
    best = fit_and_choose_model(X_tr, y_tr, X_val, y_val, idx_val, val_excess)

    # get probabilities
    p_fit, p_test = model_probabilities(best, X_train, X_test)

    # diagnostics on TEST
    auc  = roc_auc_score(y_test, p_test)
    prc  = average_precision_score(y_test, p_test)

    # TEST positions
    th, band = best["thr"], best["band"]
    pos_test = np.where(p_test >= th + band, 1, np.where(p_test <= th - band, -1, 0))
    pos_test = pd.Series(pos_test, index=idx_test)

    # Strategy returns (excess)
    r_excess = df.loc[idx_test, 'excess_ret_next']
    strat_gross = pos_test * r_excess
    costs       = txn_costs_from_positions(pos_test, PER_SIDE_COST)
    strat_net   = (strat_gross + costs).dropna()

    perf = perf_stats(strat_net)
    out = dict(
        ticker=tkr, market=market_symbol,
        sector=meta["sector"], size=meta["size"],
        best_model=best["model"],
        best_params=(best.get("C") if best["model"]=="LR" else best.get("params")),
        threshold=float(th), band=float(band),
        val_sharpe=float(best["val_sharpe"]),
        test_auc=float(auc), test_pr=float(prc),
        strat_returns=strat_net,
        pos=pos_test,
        df=df, features=features
    )
    return out

def combine_portfolio(results):
    pnl = pd.DataFrame({r['ticker']: r['strat_returns'] for r in results})
    port = pnl.mean(axis=1, skipna=True)
    return port, {r['ticker']: perf_stats(r['strat_returns']) for r in results}, perf_stats(port)

# -----------------------------
# RUN SUITES (SPY and QQQ hedges)
# -----------------------------
def run_suite(market_symbol):
    results = []
    for t in TICKERS:
        try:
            results.append(run_single_ticker(t, market_symbol))
        except Exception as e:
            print(f"[WARN] Skipping {t} on {market_symbol}: {e}")
    port, per_tkr, port_perf = combine_portfolio(results) if results else (pd.Series(dtype=float), {}, {})
    return results, port, per_tkr, port_perf

print("Running suite: MARKET = SPY (with sector + size where applicable)")
res_spy, port_spy, per_spy, port_spy_perf = run_suite(MARKET_DEFAULT)

print("Running suite: MARKET = QQQ (with sector + size where applicable)")
res_qqq, port_qqq, per_qqq, port_qqq_perf = run_suite(MARKET_ALT)

# -----------------------------
# REPORTING
# -----------------------------
def pretty_perf(p): 
    return f"CAGR={p['CAGR']:.2%} | Sharpe={p['Sharpe']:.2f} | MaxDD={p['MaxDD']:.1%}"

print("\n=== PER-TICKER (SPY + Sector + Size) ===")
for r in res_spy:
    stats = perf_stats(r['strat_returns'])
    print(f"{r['ticker']:>6s}  [{r['best_model']}, thr={r['threshold']:.2f}, band={r['band']:.2f}]  "
          f"AUC={r['test_auc']:.3f}, PR={r['test_pr']:.3f}  |  "
          f"Sector={r['sector']}, Size={r['size']}  ||  {pretty_perf(stats)}")

print("\n=== PORTFOLIO (Equal-Weight) — SPY suite ===")
print(pretty_perf(port_spy_perf))

print("\n=== PER-TICKER (QQQ + Sector + Size) ===")
for r in res_qqq:
    stats = perf_stats(r['strat_returns'])
    print(f"{r['ticker']:>6s}  [{r['best_model']}, thr={r['threshold']:.2f}, band={r['band']:.2f}]  "
          f"AUC={r['test_auc']:.3f}, PR={r['test_pr']:.3f}  |  "
          f"Sector={r['sector']}, Size={r['size']}  ||  {pretty_perf(stats)}")

print("\n=== PORTFOLIO (Equal-Weight) — QQQ suite ===")
print(pretty_perf(port_qqq_perf))

# -----------------------------
# PLOTS
# -----------------------------
plt.figure(figsize=(12,5))
if len(port_spy) > 0:
    (1 + port_spy).cumprod().plot(label="Portfolio: SPY + Sector(+Size)", lw=2)
if len(port_qqq) > 0:
    (1 + port_qqq).cumprod().plot(label="Portfolio: QQQ + Sector(+Size)", lw=2, alpha=0.8)
plt.title("Cumulative Returns — Portfolio (TEST segment within series)")
plt.legend()
plt.tight_layout()
plt.savefig("./figures/port_v2.png", dpi=300, bbox_inches="tight") 
plt.show()

# Per-ticker cumulative (SPY suite)
if res_spy:
    plt.figure(figsize=(12,6))
    for r in res_spy:
        (1 + r['strat_returns']).cumprod().plot(label=f"{r['ticker']}", alpha=0.85)
    plt.title("Cumulative Returns — Per Ticker (SPY suite)")
    plt.legend()
    plt.tight_layout()
    plt.savefig("./figures/stock_v2.png", dpi=300, bbox_inches="tight") 
    plt.show()
```
